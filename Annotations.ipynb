```python
import pandas as pd
```

```python
data_path = '/content/drive/MyDrive/MM/tvsum/data/ydata-tvsum50-anno.tsv'

data = pd.read_csv(data_path,sep='\t', header = None)

data.columns = ["video id", "video category", "annotations"]

# data.columns
annotations = data.loc[data["video id"] == "98MoyGZKHXc"]["annotations"].tolist()

print(len(annotations[1]))
```

```python
def aggregations(vid_id):
  '''
  aggregates frame wise annotations scores from multiple annotators
  '''

  vid_cat = data.loc[data['video id'] ==vid_id]
  annotations = vid_cat['annotations'].tolist()
  anno_list = []

  for anno in annotations:
    make_list = [x.split() for x in anno]
    anno_list.append(make_list)

  aggegation = []
  i = 0 
  while i < len(anno_list[0]):
    sum = 0
    for j in range(len(anno_list)):
      sum += int(anno_list[j][i][0])
    i += 2
    aggegation.append(sum/len(anno_list))

  return aggegation
```

```python
total_agg = aggregations('98MoyGZKHXc') # aggregate annotations

results = []

# code to adapt the annotations to the current frame rate

frames = 179
divided = 26
for i in range(frames):
  score = 0
  for j in range(divided):
    score += total_agg[i*divided + j]
  score = score/divided
  results.append(score)

print(results)

annotations = {'video_id' : {}, 'labels' : {}}
annotations['video_id'] = '98MoyGZKHXc'
annotations['labels'] = results

annotations = pd.DataFrame(annotations)

annotations.to_csv('/content/drive/MyDrive/MM/temp_features/annotations.csv', index=None)  # save aggregated annotations to csv
```
